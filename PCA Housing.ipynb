# PCA Using Sklearn

```python
from sklearn.decomposition import PCA as sklearnPCA
```

We have 321 columns. So first we choose 319 components

```python
# create instance for pca
sklearn_pca = sklearnPCA(n_components = 319)

#fit and transform the instance on data
y_sklearn = sklearn_pca.fit_transform(x_std)
```

```python
print(sklearn_pca.explained_variance_ratio_.cumsum()) #it is showing that how much variance my each component of variance
#is showingand it is cumulative sum so first value is of 1st variable and 2nd value is of 1st+2nd variable and so on
```

```python
count = 0
for i in range(len(sklearn_pca.explained_variance_ratio_.cumsum())):
    if sklearn_pca.explained_variance_ratio_.cumsum()[i] <= .9:
        count+= 1

print("No. of Components: "+ str(count))        
```

We have 90% variance with only 192 components from 319 of total components 

```python
plt.figure()
cum_sum = np.cumsum(sklearn_pca.explained_variance_ratio_)
plt.scatter([200],[cum_sum[200]],color = 'r')
plt.plot(np.cumsum(sklearn_pca.explained_variance_ratio_))
plt.xlabel('Principal Component')
plt.ylabel('Variance Ratio(%)') #for each component
plt.title('Dataset Explained Variance')
plt.show()
```
